# Customer Segmentation Using Clustering (Python)

# IMPORT LIBARIES

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.decomposition import PCA
```

# LOAD DATASET

```python
data = pd.read_csv(r"C:\Users\vinee\Downloads\Zaalima Python Project\telecom_churn.csv")
data
```

```python
data.head()
```

# DROP UNWANTED COLUMNS

```python
data.drop([
    'customer_id',
    'telecom_partner',
    'state',
    'city',
    'pincode',
    'date_of_registration',
    'churn'
], axis=1, inplace=True)
```

# HANDLING MISSING VALUES

```python
data.isnull().sum()
```

```python
data = data.replace(" ", np.nan)
data = data.dropna()
```

```python
data.info()
```

# SEPARATE CATEGORICAL AND NUMERICAL

```python
categorical_cols = ['gender']
numeric_cols = ['age', 'num_dependents', 'estimated_salary',
                'calls_made', 'sms_sent', 'data_used']
```

```python
data
```

# ENCODE CATEGORICAL VARIABLES

```python
le = LabelEncoder()
data['gender'] = le.fit_transform(data['gender'])
```

```python
data
```

# SCALE NUMERICAL COLUMNS

```python
scaler = StandardScaler()
df_scaled = scaler.fit_transform(data)
```

# ELBOW METHOD

```python
inertia = []
K = range(2, 11)

for k in K:
    km = KMeans(n_clusters=k, random_state=42)
    km.fit(df_scaled)
    inertia.append(km.inertia_)

plt.plot(K, inertia, marker='o')
plt.title("Elbow Method")
plt.xlabel("Number of clusters")
plt.ylabel("Inertia")
plt.show()
```

# SILHOUETTE SCORE

```python
for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=42).fit(df_scaled)
    score = silhouette_score(df_scaled, kmeans.labels_)
    print(f"Clusters={k} → Silhouette Score={score}")
```
